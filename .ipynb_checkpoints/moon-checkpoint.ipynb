```python
import skimage
import skimage.io
import skimage.color
import skimage.measure
import skimage.feature
import skimage.filters
import skimage.transform
import pytesseract
import imutils
import matplotlib.pyplot as plt
import numpy as np

pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"
img = skimage.io.imread("2048-2.png")
img = skimage.color.rgb2gray(img[:,:,0:3])
h = 800
w = 800
img = skimage.transform.resize(img, (h, w))

plt.rcParams['figure.figsize'] = [4, 4]
plt.imshow(img, cmap="gray", vmin=0, vmax=1.0)
```

```python

plt.rcParams['figure.figsize'] = [5, 5]
fig, axs = plt.subplots(nrows=4, ncols=4, sharex=True, sharey=True)

numbers = np.zeros((4,4), dtype="int")

for i in range(4):
    for j in range(4):
            
        h = img.shape[0] 
        w = img.shape[1]
        
        y1 = int(h/4.0*i)
        y2 = int(h/4.0*(i+1))
        x1 = int(w/4.0*j)
        x2 = int(w/4.0*(j+1))
            
        crop = img[y1:y2, x1:x2]
        edge = skimage.feature.canny(crop, sigma=2.0, high_threshold=0.01, low_threshold=0.005)
        
        for k in range(0, edge.shape[1], 1):
            if(edge[edge.shape[0]//2,k]):
                y1 = k+crop.shape[0]//20
                break
        for k in range(0, edge.shape[0], 1):
            if(edge[k, edge.shape[0]//2]):
                x1 = k+crop.shape[1]//20
                break
        for k in range(edge.shape[1]-1, -1, -1):
            if(edge[edge.shape[0]//2,k]):
                y2 = k-crop.shape[0]//20
                break   
        for k in range(edge.shape[0]-1, -1, -1):
            if(edge[k, edge.shape[0]//2]):
                x2 = k-crop.shape[1]//20
                break
        
        thresh = skimage.filters.threshold_otsu(crop)
        crop = (crop > thresh).astype("float64")
        crop = crop[y1:y2, x1:x2]
        
        if(np.mean(crop) < 0.5):
            crop = 1.0 - crop
        
        axs[i,j].imshow(crop, cmap="gray", vmin=0.0, vmax=1.0)
        
        number = pytesseract.image_to_string((crop*255.0).astype("uint8"), config=' --psm 10 --oem 3 -c tessedit_char_whitelist=0123456789')
        if(number != ""):
            numbers[i,j] = int(number)
        
plt.xticks([])
plt.yticks([])
plt.show()

print(numbers)

```
